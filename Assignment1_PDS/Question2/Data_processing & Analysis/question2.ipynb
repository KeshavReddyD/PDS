```python
import pandas as pd
student_raw_data = pd.read_csv("/content/StudentsPerformance.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for Math Scores
sns.histplot(student_data_cleaned['math score'], kde=True, color='blue')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.stripplot(x="race/ethnicity", y="math score", data=student_data_cleaned, hue="race/ethnicity", jitter=True, palette="Set1", dodge=True, legend=False)
plt.xlabel("Race/Ethnicity")
plt.ylabel("Math Score")
plt.title("Math Scores by Ethnicity")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Bar Plot
sns.barplot(x="gender", y="math score", data=student_data_cleaned, errorbar=None, hue="gender", dodge=False, palette="pastel")
plt.title("Average Math Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Math Score")
plt.legend([],[], frameon=False)  # Removes redundant legend
plt.show()
```

```python
# Histogram
plt.hist(student_data_cleaned["reading score"], bins=10, color="skyblue", edgecolor="black")
plt.title("Distribution of Reading Scores")
plt.xlabel("Reading Score")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x="gender", y="writing score", hue="gender", data=student_data_cleaned, palette="coolwarm", legend=False)
plt.title("Writing Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Writing Score")
plt.show()
```
